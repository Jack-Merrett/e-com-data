```python
# install the proper libraries
import pandas as pd
```

```python
# establishing the data sets as individual varables
data_cust = pd.read_csv('../e-com-data/raw_data/CSVs/olist_customers_dataset.csv')

data_geo = pd.read_csv('../e-com-data/raw_data/CSVs/olist_geolocation_dataset.csv')

data_order_item = pd.read_csv('../e-com-data/raw_data/CSVs/olist_order_items_dataset.csv')

data_payment =  pd.read_csv('../e-com-data/raw_data/CSVs/olist_order_payments_dataset.csv')

data_reviews = pd.read_csv('../e-com-data/raw_data/CSVs/olist_order_reviews_dataset.csv')

data_orders = pd.read_csv('../e-com-data/raw_data/CSVs/olist_orders_dataset.csv')

data_product = pd.read_csv('../e-com-data/raw_data/CSVs/olist_products_dataset.csv')

data_sellers = pd.read_csv('../e-com-data/raw_data/CSVs/olist_sellers_dataset.csv')

data_product_catagory_name_trans = pd.read_csv('../e-com-data/raw_data/CSVs/product_category_name_translation.csv')
```
